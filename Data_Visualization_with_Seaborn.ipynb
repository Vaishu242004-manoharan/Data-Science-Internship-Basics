```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Iris dataset
iris = sns.load_dataset("iris")

# Histograms for numerical columns
iris.hist(figsize=(10, 6), bins=20, edgecolor='black')
plt.suptitle("Histograms of Numerical Columns", fontsize=14)
plt.show()

# Box plots for numerical columns
plt.figure(figsize=(10, 6))
for i, column in enumerate(iris.select_dtypes(include=['float64']).columns, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=iris[column])
    plt.title(f'Boxplot of {column}')
plt.tight_layout()
plt.show()

# Correlation heatmap
plt.figure(figsize=(8, 6))
corr_matrix = iris.select_dtypes(include=['float64']).corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
